```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import csv
import cx_Oracle
```

```python
### 1번 '빅데이터' 검색 후 정의 내용 txt파일 저장하기

file = open('data/빅데이터.txt', 'w', encoding='utf-8')

driver = webdriver.Chrome()
url = 'https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%8C%80%EB%AC%B8'
driver.get(url)
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="p-search"]/a/span[1]').click()
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="searchform"]/div/div/div[1]/input').send_keys('빅데이터')
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="searchform"]/div/button').click()
time.sleep(2)
data = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/p[6]').text
print(data)
file.write(data)

driver.close()
```
