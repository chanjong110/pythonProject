```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import csv
```

```python
### 1번 '빅데이터' 검색 후 정의 내용 콘솔 출력하기
driver = webdriver.Chrome()
url = 'https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%8C%80%EB%AC%B8'
driver.get(url)
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="p-search"]/a/span[1]').click()
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="searchform"]/div/div/div[1]/input').send_keys('빅데이터')
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="searchform"]/div/button').click()
time.sleep(2)
data = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/p[6]').text
print(data)

driver.close()
```

```python
### 2번 동의 진행 중인 청원 내용 수집하여 '국민동의.txt' 파일로 저장

file = open('data/국민동의.txt', 'w', encoding='utf-8')

driver = webdriver.Chrome()
url = 'https://petitions.assembly.go.kr/'
driver.get(url)
time.sleep(2)
driver.find_element(By.XPATH, '//*[@id="mainMenuList_1"]').click()
time.sleep(2)
p_cnt = int(driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li[9]/a').text)
time.sleep(2)
driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li[3]/a').click()
for i in range(p_cnt):
    mpage =  len(driver.find_elements(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li'))-1
    list_preader = driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li['+ str(mpage) +']/a').click() 
    time.sleep(2)
    tit_cont = len(driver.find_elements(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[1]/div[2]/ul/li'))+1
    for i in range(1, tit_cont):
        time.sleep(1)
        tit = driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[1]/div[2]/ul/li['+ str(i) +']/div/a/dl/dd').text
        file.write(tit+'\n')
        print(tit)
        
        
file.close()
driver.close()
```

```python
### 3번 셀레니움 라이브러리 활용하여 대전 서구 주유소 보통 희발유 가격을 기록한 엑셀 파일 저장

driver = webdriver.Chrome()
url = 'https://www.opinet.co.kr/user/main/mainView.do'
driver.get(url)

time.sleep(2)
driver.find_element(By.XPATH,'//*[@id="quick_ul"]/li[2]/a').click()
time.sleep(3)
driver.find_element(By.XPATH,'//*[@id="searRgSelect"]').click()
time.sleep(2)
file = open('data/oilinfo.csv', 'w', newline='')
writer = csv.writer(file)

oil_rows = driver.find_elements(By.XPATH,'//*[@id="body1"]/tr')
for i in range(1, len(oil_rows)+1):
    oil_name = driver.find_element(By.XPATH,'//*[@id="body1"]/tr['+ str(i) +']/td[1]/a').text
    g_price = driver.find_element(By.XPATH,'//*[@id="body1"]/tr['+ str(i) +']/td[2]').text
    v_price = driver.find_element(By.XPATH,'//*[@id="body1"]/tr['+ str(i) +']/td[3]').text
    # print(oil_name, g_price, v_price)
    writer.writerow(['대전 서구', oil_name, g_price, v_price])

file.close()
driver.close()
```
